```python
from pathlib import Path
from zipfile import ZipFile
import json
import pandas as pd
```

```python
photos_dataset = []

with ZipFile(ACCOUNT_PART_ONE_ZIP_PATH, "r") as zip_file_obj:
    file_names = [
        name for name in zip_file_obj.namelist()
        if name.startswith("photo_") is True
    ]

    for file_name in file_names:
        with zip_file_obj.open(file_name) as file_obj:
            photos_dataset.append(json.loads(file_obj.read()))

df_photos = pd.DataFrame(photos_dataset)
df_photos.set_index("id", inplace=True)
df_photos.head()
```

```python
ACCOUNT_PART_ONE_ZIP_PATH = Path("data/source/72157721559577957_ed7cb68ec9ab_part1.zip").resolve()

account_datasets = {}

with ZipFile(ACCOUNT_PART_ONE_ZIP_PATH, "r") as zip_file_obj:
    file_names = [
        name for name in zip_file_obj.namelist()
        if name.startswith("photo_") is False
    ]

    for file_name in file_names:
        with zip_file_obj.open(file_name) as file_obj:
            account_datasets[file_name] = json.loads(file_obj.read())

for key in account_datasets.keys():
    print(key)
```
